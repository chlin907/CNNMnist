## Use 3 image files to test the flow 

```python
from CNNMnist.CNNMnist import CNNMnist
import numpy as np

from IPython import display

model_path = './weights.04-0.07.hdf5'

cnn_mnist = CNNMnist()
cnn_mnist.load_model(model_path)

img_path = './doc/test_2.png'
proba = cnn_mnist.predict_proba(img_path)
print('Number is {} with probability = {:.2f}%'.format(np.argmax(proba), 100 * np.max(proba)))
display.Image(img_path)
```
