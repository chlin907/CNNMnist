## Review image data for all kinds of digits

```python
num_fig_per_num = 10

x_train, y_train = cnn_mnist.get_train_set()
x_train = x_train[:, :, :]


num_idx_map = {num:[] for num in range(10)}
for idx, num in enumerate(y_train):
    if sum([sum(v) for v in num_idx_map.values()]) == num_fig_per_num * 10:
        break
    if len(num_idx_map[num]) < num_fig_per_num:
        num_idx_map[num].append(idx)

fig, axes = plt.subplots(10, num_fig_per_num, figsize=(15,15))
for i in range(10):
    for j in range(num_fig_per_num):
        sns.heatmap(x_train[num_idx_map[i][j]], cmap="Greys", cbar=False, square=True,
                    xticklabels=False, yticklabels=False, ax=axes[i,j])
```
