## Pipeline from input dataset folder to trained model

```python
from CNNMnist.CNNMnist import CNNMnist

path = './data'
cnn_mnist = CNNMnist(path)
cnn_mnist.load_dataset()

cnn_mnist.preprocess()
cnn_mnist.build_model()
cnn_mnist.fit(num_epoch=20, num_batch_size=64, val_split=0.2)
test_loss, test_acc = cnn_mnist.test()

print('Test result: loss = {} and acc = {}'.format(test_loss, test_acc))
```

## Training and validation loss (cross entropy) and accuracy versus epoch

```python
import matplotlib.pyplot as plt
train_acc = cnn_mnist.history.history['acc']
val_acc = cnn_mnist.history.history['val_acc']
train_loss = cnn_mnist.history.history['loss']
val_loss = cnn_mnist.history.history['val_loss']

epochs = range(1, len(train_acc) + 1)

plt.plot(epochs, train_acc, 'bo-', label = 'Training acc')
plt.plot(epochs, val_acc, 'ro-', label = 'Validation acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.figure()

plt.plot(epochs, train_loss, 'bo-', label = 'Training loss')
plt.plot(epochs, val_loss, 'ro-', label = 'Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```
